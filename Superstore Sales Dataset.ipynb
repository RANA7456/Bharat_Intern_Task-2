```python
top_states = df.groupby("State").sum().sort_values("Sales", ascending=False).head(15)
top_states = top_states[["Sales"]].round(2)
top_states.reset_index(inplace=True)
print(top_states)
```

```python
plt.figure(figsize=(17,7))
plt.bar(top_states["State"], top_states["Sales"])
plt.title("Top 15 States with Sales",fontsize=25)
plt.xlabel("States", fontsize=20)
plt.ylabel("Sales", fontsize=20)
plt.xticks(rotation='vertical')
for k,v in top_states["Sales"].items():
    if v > 100000:
        plt.text(k,v-100000, f"$ {str(v)}", rotation=90)
    else:
        plt.text(k,v-30000, f"$ {str(v)}", rotation=90)
```
